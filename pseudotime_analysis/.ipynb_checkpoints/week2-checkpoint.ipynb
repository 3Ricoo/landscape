## Dimensionality reductions


### SC3

![CS3 pipline](pic/sc3.png)
__SC3 [(Kiselev et al, 2017)](https://www.nature.com/articles/nmeth.4236) is based on PCA and spectral dimensionality reductions with utilizing k-means.__


__SC3 can estimate the number of clusters__

```r
deng <- sc3_estimate_k(deng)
```

```r
metadata(deng)$sc3$k_estimation
```

The number of cell types predicted by SC3 is smaller than in the original data annotation. However, early, mid and late stages of different cell types together, we will have exactly 6 cell types.

```r
plotPCA(deng, colour_by = "cell_type1")
```

```r
deng <- sc3(deng, ks = 10, biology = TRUE, n_cores = 1)
```

```r
sc3_plot_consensus(deng, k = 10, show_pdata = "cell_type2")
```

__Fig: Consensus matrix__

```r
sc3_plot_expression(deng, k = 10, show_pdata = "cell_type2")
```

__Fig: Heatmap of the expression matrix:__

```r
sc3_plot_markers(deng, k = 10, show_pdata = "cell_type2")
```

__Fig: Identified marker genes__

```r
plotPCA(deng, colour_by = "sc3_10_clusters")
```

__Fig: PCA plot with highlighted SC3 clusters__

__Adjusted Rand index(ARI):__ Compare the results of SC3 clustering with the original publication cell type labels:

```r
adjustedRandIndex(colData(deng)$cell_type2, colData(deng)$sc3_10_clusters)
```

### tSNE + kmeans

```r
deng <- runTSNE(deng, rand_seed = 1)
plotTSNE(deng)
```

```r
colData(deng)$tSNE_kmeans <- as.character(kmeans(deng@reducedDims$TSNE, centers = 10)$clust)
plotTSNE(deng, colour_by = "tSNE_kmeans")
```

tSNE+kmeans is __stochastic__ and gives different results every time they are run. To get a better overview of the solutions, we need to run the methods multiple times. SC3 is also stochastic, yet due to the consensus step, it is more robust and less likely to produce different outcomes.
